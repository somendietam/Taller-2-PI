**Pregunta 1**

Utilice histogramas para comparar la distribución de la columna _imdb_score_ relacionada con películas producidas en Francia y Canadá.

Para crear el histograma:

- Ajuste el número de _bins_ a 20
- De nombres a las etiquetas de los ejes (xAxis y yAxis)
- Remueva el _grid_

```python
df_canada = df[df['country'] == 'Canada']
fig = px.histogram(df_canada, x="imdb_score",nbins=20,labels={'imdb_score':'score_canada'})
fig.show()

df_france = df[df['country'] == 'France']
fig = px.histogram(df_france, x="imdb_score",nbins=20,labels={'imdb_score':'score_france'})
fig.show()
```

**Pregunta 3**

¿Qué país tiene el mayor promedio de calificación de películas? Utilice figuras de barras para visualizar los resultados.  



```python
country_score = df.groupby('country')['imdb_score'].mean()
print(country_score)

print(f"El pais con mayor promedio de calificación fue {country_score.idxmax()} con {country_score.max()} de calificación")

fig = px.bar(x = country_score.index, y = country_score, labels={'y':'Calificacion', 'x':'Paises'})
fig.show()
```

**Pregunta 5**

¿Cuál película de Christopher Nolan tiene la mayor calificación (_imdb_score_)? Utilice figuras de barras para visualizar los resultados.

```python
director = df[df['director_name'] == 'Christopher Nolan']
mayor = director['imdb_score'].idxmax()

print(f"La película con mayor calificacion de Christopher Nolan es: {df.iloc[mayor]['movie_title']} con una calificación de: {director['imdb_score'].max()}")
fig = px.bar(x = director['movie_title'], y = director['imdb_score'], labels={'y':'Calificacion', 'x':'Paises'})
fig.show()


```
